```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier

""" parsed_txt_file = pd.read_csv(r'./diabetes.txt')
parsed_txt_file.to_csv(r'./diabetes.csv',index=None) """
datos = pd.read_csv('./diabetes.csv', header=0)

edad = list(datos['AGE'])
genero = list(datos['SEX'])
peso = list(datos['BMI'])

data = {'Masa': peso,
        'Edad': edad,
        'Genero': genero}
punto_nuevo = {'Masa': [39],
               'Edad': [33]}
df = pd.DataFrame(data)
punto_nuevo = pd.DataFrame(punto_nuevo)
# sns.scatterplot(df['Masa'], df['Edad'], hue=df['sex'])
ax = plt.axes()
ax.scatter(df.loc[df['Genero'] == 1, 'Masa'],
           df.loc[df['Genero'] == 1, 'Edad'],
           c="red",
           label="Hombre")
ax.scatter(df.loc[df['Genero'] == 2, 'Masa'],
           df.loc[df['Genero'] == 2, 'Edad'],
           c="blue",
           label="Mujer")
ax.scatter(punto_nuevo['Masa'],
           punto_nuevo['Edad'],
           c="black")
plt.xlabel("Masa")
plt.ylabel("Edad")
ax.legend()
plt.show()

knn = KNeighborsClassifier(n_neighbors=3)
X = df[['Masa', 'Edad']]
y = df[['Genero']]
knn.fit(X, y)
prediccion = knn.predict(punto_nuevo)
if (prediccion == 1):
    print(f'{prediccion} | Hombre')
else:
    print(f'{prediccion} | Mujer')
```
